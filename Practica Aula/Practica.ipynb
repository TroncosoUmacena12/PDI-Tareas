Laplaciano 

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('Cespet.jpg', cv2.IMREAD_GRAYSCALE)

laplaciano = cv2.Laplacian(img, cv2.CV_64F)

laplaciano_visible = cv2.convertScaleAbs(laplaciano)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Laplaciano (Visible)")
plt.imshow(laplaciano_visible, cmap='gray')
plt.axis('off')
plt.show()
```

Laplaciano del Gaussiano

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("Cespet.jpg", cv2.IMREAD_GRAYSCALE)

gauss = cv2.GaussianBlur(img, (5, 5), 0)

laplaciano = cv2.Laplacian(gauss, cv2.CV_64F)

laplaciano_visible = cv2.convertScaleAbs(laplaciano)


plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Gaussiana")
plt.imshow(gauss, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("LoG (Visible)")
plt.imshow(laplaciano_visible, cmap='gray')
plt.axis('off')

plt.show()
```
